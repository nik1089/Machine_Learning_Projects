```python
# Although in case of KNN outliers do not affect the model, yet lets try outlier treatment and then see the diiference in results
# Let's visualize all the outliers present in data again

plt.figure(figsize=(15, 12))

for i, variable in enumerate(columns):
    plt.subplot(4, 4, i + 1)
    plt.boxplot(df_all[variable], whis=1.5)
    plt.tight_layout()
    plt.title(variable)
```

```python
numeric_columns = ['P_incidence', 'P_tilt', 'L_angle', 'S_slope', 'P_radius', 'S_Degree']
```

```python
# to find the 25th percentile and 75th percentile for the columns.
Q1 = df_all[numeric_columns].quantile(0.25)
Q3 = df_all[numeric_columns].quantile(0.75)

IQR = Q3 - Q1                   #Inter Quantile Range (75th percentile - 25th percentile)

lower_whisker = Q1 - 1.5*IQR    #Finding lower and upper bounds for all values. All values outside these bounds are outliers
upper_whisker = Q3 + 1.5*IQR
```

```python
# Percentage of outliers in each column
((df_all[numeric_columns] < lower_whisker) | (df_all[numeric_columns] > upper_whisker)).sum()/df_all.shape[0]*100
```

```python
def treat_outliers(df, col):
    """
    treats outliers in a variable
    col: str, name of the numerical variable
    df: dataframe
    col: name of the column
    """
    Q1 = df[col].quantile(0.25)  # 25th quantile
    Q3 = df[col].quantile(0.75)  # 75th quantile
    IQR = Q3 - Q1                # Inter Quantile Range (75th perentile - 25th percentile)
    lower_whisker = Q1 - 1.5 * IQR
    upper_whisker = Q3 + 1.5 * IQR

    # all the values smaller than lower_whisker will be assigned the value of lower_whisker
    # all the values greater than upper_whisker will be assigned the value of upper_whisker
    # the assignment will be done by using the clip function of NumPy
    df[col] = np.clip(df[col], lower_whisker, upper_whisker)

    return df
```

```python
df_all = treat_outliers(df_all, 'P_incidence')
# visualizing the column after outlier treatment
sns.boxplot(data=df_all, x='P_incidence');
```

```python
# Similarly we will treat the outliers in other columns using the same approach
df_all = treat_outliers(df_all, 'P_tilt')
df_all = treat_outliers(df_all, 'L_angle')
df_all = treat_outliers(df_all, 'S_slope')
df_all = treat_outliers(df_all, 'P_radius')
df_all = treat_outliers(df_all, 'S_Degree')

plt.figure(figsize=(15, 12))

for i, variable in enumerate(columns):
    plt.subplot(4, 4, i + 1)
    plt.boxplot(df_all[variable], whis=1.5)
    plt.tight_layout()
    plt.title(variable)
```

```python
# blancing data with SMOTE
from collections import Counter
from imblearn.over_sampling import SMOTE

counter = Counter(y)
print('Before Counter', counter)

smt = SMOTE(sampling_strategy='auto', random_state=2)

X_resampled, y_resampled = smt.fit_resample(X, y)

after_counter = Counter(y_resampled)
print('After Counter', after_counter)
```
